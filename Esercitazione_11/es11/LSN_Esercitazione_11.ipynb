### <span style="color: #FF5733 "> Esercizio 11.2</span>

Ho provato a estendere il modello per fittare la funzione:$$
f(x)=4-3x-2x^2+3x^3
$$
con $x \in [-1,1]$.  

Per tutti i modelli ho usato $N_{train}=1000$ e ho tenuto l'ottimizzatore SGD.  

Si riportano di seguito il fit prodotto nell'intervallo di training (a sinistra) e il fit in un intervallo più ampio (a destra).  


**Primo tentativo:** Ho usato $N_{epoche}=50$ e ho realizzato un DNN con:  
- un layer da 1 neurone con attivazione spftplus
- un layer da 1 neurone con attivazione spftplus
- un layer da 1 neurone con attivazione softplus

```python
a = 4
b = -3
c = -2
d = 3
x_valid = np.random.uniform(-1, 1, 100)
x_valid.sort()
y_target = a+b*x_valid+c*x_valid*x_valid+d*x_valid*x_valid*x_valid
x_valid_out = np.random.uniform(-2, 2, 100)
x_valid_out.sort()
y_target_out = a+b*x_valid_out+c*x_valid_out*x_valid_out+d*x_valid_out*x_valid_out*x_valid_out
model=load_model("Modelli_2/Modello_1.mod")
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)
print('Test accuracy model_1:', score[1])
x_predicted = np.random.uniform(-1, 1, 100)
x_outrange = np.random.uniform (-2,2,100)
y_outrange =model.predict(x_outrange)
y_predicted = model.predict(x_predicted)

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)

plt.subplot(1,2,2)
plt.scatter(x_outrange, y_outrange,color='r')
plt.plot(x_valid_out, y_target_out)
plt.grid(True)

plt.show()
```

**Secondo tentativo:** Ho provato a modificare un'attivazione a aumentare il numero di neuroni, per aumentare il numero di paramentri. Ho usato $N_{epoche}=100$ e ho realizzato un DNN con:  
- un layer da 10 neuroni con attivazione spftplus
- un layer da 10 neuroni con attivazione sigmoid
- un layer da 1 neurone con attivazione softplus

```python
model=load_model("Modelli_2/Modello_2.mod")
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)
print('Test accuracy model_2:', score[1])
x_predicted = np.random.uniform(-1, 1, 100)
x_outrange = np.random.uniform (-2,2,100)
y_outrange =model.predict(x_outrange)
y_predicted = model.predict(x_predicted)

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)

plt.subplot(1,2,2)
plt.scatter(x_outrange, y_outrange,color='r')
plt.plot(x_valid_out, y_target_out)
plt.grid(True)

plt.show()
```
