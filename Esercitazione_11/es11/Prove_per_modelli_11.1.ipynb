```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect
```

```python
import numpy as np

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 1000) 
x_valid = np.random.uniform(-1, 1, 100)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

sigma = 0.05# noise standard deviation, for the moment it is absent
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)
```

```python
# plot training and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
#plt.scatter(x_train, y_train, color='g')
plt.grid(True); plt.show()
```

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model_1 = tf.keras.Sequential()
model_1.add(Dense(1, input_shape=(1,))) #ho una sola ascissa come input

# compile the model choosing optimizer, loss and metrics objects
model_1.compile(optimizer='sgd', loss='mse', metrics=['mse']) #ho un solo layer con un neurone
#mse=nean square error
#in questo caso semplice, ha tutto per poter lavorare
```

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model_1.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=500,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# return weights
from tensorflow.keras.models import load_model
model_1.get_weights()
save_model_path='Modelli/Modello_16.mod'
model_1.save(filepath=save_model_path, include_optimizer=True)
```

```python
# evaluate model
score = model_1.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
score = model_1.evaluate(x_train, y_train, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model_1.evaluate(x_valid, y_target, batch_size=32, verbose=1)
# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python

x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model_1.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.title("$\sigma=0.3, N_{epochs}=50, N_{train}=500$ ")
plt.grid(True)
plt.show()
```
